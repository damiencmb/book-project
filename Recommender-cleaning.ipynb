# Pages

The type of 'pages' column is 'object' because some rows contain values like '1page' so we first need to change it as interger and then we replace the NaN values by the mean.

```python
# we first need to keep only the numeric part 
df['pages'] = df['pages'].str.extract('(\d+)')
```

```python
# we convert to interger
df['pages'] = df['pages'].astype(int)
```

# Price

```python
# df['price'] = df['price'].astype(float)
```

I got this message error : ValueError: could not convert string to float: '1.189.88'


It seems that that this price book is not in dollar but we can not guess which currency it is so we check how many row have this unusal price format and drop it if there are not too many.

```python
# we look for the unsual price pattern
pattern = '^\d+\.\d+\.\d+$'

# then we check the number of rows with this price pattern
rowspattern = df['price'].str.match(pattern)
rowspattern.value_counts()
```

```python
# We create a new boolean column 
df['boolprice'] = rowspattern

# We check of the unsual prices fomart are there
rowstrue = df[(df['boolprice']==True)]
rowstrue.head()
```

```python
# we drop the unsual pattern rows
df = df[df['boolprice'] != True]
```

```python
# we drop the column 'boolprice'
df = df.drop(columns=['boolprice'])

# we convert the 'price' column into float
df['price'] = df['price'].astype(float)
```

```python
df.head()
```

# Ratings by stars


We need to seperate the column ratingsByStars in 5, one for each number of stars ratings.

```python
# we first remove the apostrophes and square brackets
df['ratingsByStars'] = [e.replace("'", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("[", "") for e in df['ratingsByStars']]
df['ratingsByStars'] = [e.replace("]", "") for e in df['ratingsByStars']]
df.head()
```

```python
# then we split the values on the coma
df[['5*', '4*', '3*', '2*', '1*']] = df['ratingsByStars'].str.split(',', expand=True)
df = df.drop(columns=['ratingsByStars'])
```

```python
df.isna().sum()
```

We suppose that NaN values for star ratings are because people only gave a 5 stars rating so we replace them by 0.

```python
df[['5*', '4*', '3*', '2*', '1*']] = df[['5*', '4*', '3*', '2*', '1*']].fillna(0)
df.isna().sum()
```
